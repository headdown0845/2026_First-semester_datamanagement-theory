```python
u = np.array([3, 5]); v = np.array([6, 2])
dot_product = np.dot(u, v)
norm_u = np.linalg.norm(u); norm_v = np.linalg.norm(v)
```

```python
# Calculate the cosine of the angle
cos_theta = dot_product / (norm_u * norm_v)
print(f"Vector u and v: {u} and {v}")
print(f"Dot product (u · v): {dot_product}")
print(f"Norm of u (||u||) and v (||v||): {norm_u:.2f} & {norm_v:.2f}")
print(f"Cosine of the angle (cos(theta)): {cos_theta:.4f}")
print(f"Result of norm_u*norm_v*cos_theta: {norm_u * norm_v * cos_theta:.2f}")
```

```python
def T(v):
    w = np.zeros((3,1))
    w[0,0] = 3*v[0,0]
    w[2,0] = -2*v[1,0]

    return w

v = np.array([[3], [5]])
w = T(v)

print("Original vector:\n", v, "\n\n Result of the transformation:\n", w)
```

```python
u = np.array([[1], [-2]])
v = np.array([[2], [4]])

k = 7

print("T(k*v):\n", T(k*v), "\n\n k*T(v):\n", k*T(v), "\n\n")
print("T(u+v):\n", T(u+v), "\n\n T(u)+T(v):\n", T(u)+T(v))
```

```python
def L(v):
    A = np.array([[3,0], [0,0], [0,-2]])
    print("Transformation matrix:\n", A, "\n")
    w = A @ v

    return w

v = np.array([[3], [5]])
w = L(v)

print("Original vector:\n", v, "\n\n Result of the transformation:\n", w)
```

```python
def T_hscaling(v):
    A = np.array([[2,0], [0,1]])
    w = A @ v

    return w


def transform_vectors(T, v1, v2):
    V = np.hstack((v1, v2))
    W = T(V)

    return W

e1 = np.array([[1], [0]])
e2 = np.array([[0], [1]])

transformation_result_hscaling = transform_vectors(T_hscaling, e1, e2)

print("Original vectors:\n e1= \n", e1, "\n e2=\n", e2,
      "\n\n Result of the transformation (matrix form):\n", transformation_result_hscaling)
```

```python
import matplotlib.pyplot as plt
import numpy as np # Ensure numpy is imported in the function scope or globally if not already

def plot_transformation(T, e1, e2):
    color_original = "#129cab"
    color_transformed = "#cc8933"

    _, ax = plt.subplots(figsize=(7, 7))
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.set_xticks(np.arange(-5, 5))
    ax.set_yticks(np.arange(-5, 5))

    plt.axis([-5, 5, -5, 5])
    # Fix 1: Access scalar elements e.g., e1[0,0] instead of e1[0]
    plt.quiver([0, 0],[0, 0], [e1[0,0], e2[0,0]], [e1[1,0], e2[1,0]], color=color_original, angles='xy', scale_units='xy', scale=1)
    # Fix 2: Access scalar elements e.g., e1[0,0] instead of e1[0]
    plt.plot([0, e2[0,0], e1[0,0], e1[0,0]],
             [0, e2[1,0], e2[1,0], e1[1,0]],
             color=color_original)

    # Fix 3: Corrected e_sgn calculation to produce a 1D array of floats
    e1_sgn_components = np.array([np.sign(val) if val != 0 else 1 for val in e1.flatten()])
    e1_sgn = 0.4 * e1_sgn_components
    # Fix 4: Access scalar elements e.g., e1[0,0] instead of e1[0] and e1_sgn[0] instead of e1_sgn[0][0]
    ax.text(e1[0,0]-0.2+e1_sgn[0], e1[1,0]-0.2+e1_sgn[1], f'$e_1$', fontsize=14, color=color_original)

    # Fix 5: Corrected e_sgn calculation for e2
    e2_sgn_components = np.array([np.sign(val) if val != 0 else 1 for val in e2.flatten()])
    e2_sgn = 0.4 * e2_sgn_components
    # Fix 6: Access scalar elements e.g., e2[0,0] instead of e2[0]
    ax.text(e2[0,0]-0.2+e2_sgn[0], e2[1,0]-0.2+e2_sgn[1], f'$e_2$', fontsize=14, color=color_original)


    e1_transformed = T(e1)
    e2_transformed = T(e2)

    # Fix 7: Access scalar elements e.g., e1_transformed[0,0] instead of e1_transformed[0]
    plt.quiver([0, 0],[0, 0], [e1_transformed[0,0], e2_transformed[0,0]], [e1_transformed[1,0], e2_transformed[1,0]],
               color=color_transformed, angles='xy', scale_units='xy', scale=1)
    # Fix 8: Access scalar elements e.g., e1_transformed[0,0] instead of e1_transformed[0]
    plt.plot([0,e2_transformed[0,0], e1_transformed[0,0]+e2_transformed[0,0], e1_transformed[0,0]],
             [0,e2_transformed[1,0], e1_transformed[1,0]+e2_transformed[1,0], e1_transformed[1,0]],
             color=color_transformed)

    # Fix 9: Corrected e_transformed_sgn calculation
    e1_transformed_sgn_components = np.array([np.sign(val) if val != 0 else 1 for val in e1_transformed.flatten()])
    e1_transformed_sgn = 0.4 * e1_transformed_sgn_components
    # Fix 10: Access scalar elements
    ax.text(e1_transformed[0,0]-0.2+e1_transformed_sgn[0], e1_transformed[1,0]-e1_transformed_sgn[1],
            f'$T(e_1)$', fontsize=14, color=color_transformed)

    # Fix 11: Corrected e_transformed_sgn calculation for e2
    e2_transformed_sgn_components = np.array([np.sign(val) if val != 0 else 1 for val in e2_transformed.flatten()])
    e2_transformed_sgn = 0.4 * e2_transformed_sgn_components
    # Fix 12: Access scalar elements
    ax.text(e2_transformed[0,0]-0.2+e2_transformed_sgn[0], e2_transformed[1,0]-e2_transformed_sgn[1],
            f'$T(e_2)$', fontsize=14, color=color_transformed)

    plt.gca().set_aspect("equal")
    plt.show()

plot_transformation(T_hscaling, e1, e2)
```

```python
import requests
import numpy as np
import cv2

url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTe1uVXaTVX2I6an-soBDAxR9cX1nv7eh_I3g&s'

# Download the image from the URL
response = requests.get(url)
image_data = np.array(bytearray(response.content), dtype=np.uint8)

# Read the image data using cv2.imdecode
img = cv2.imdecode(image_data, cv2.IMREAD_COLOR) # Read in color (BGR by default)

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb) # Display in RGB
```

```python
image_rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(image_rotated)
```

```python
rows,cols, _ = image_rotated.shape # Unpack all three dimensions, ignoring the third one
# 3 by 3 matrix as it is required for the OpenCV library, don't worry about the details of it for now.
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
image_rotated_sheared = cv2.warpPerspective(image_rotated, M, (int(cols), int(rows)))
plt.imshow(image_rotated_sheared)
```

```python
image_sheared = cv2.warpPerspective(img, M, (int(cols), int(rows)))
image_sheared_rotated = cv2.rotate(image_sheared, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(image_sheared_rotated)
```

```python
M_rotation_90_clockwise = np.array([[0, 1], [-1, 0]])
M_shear_x = np.array([[1, 0.5], [0, 1]])

print("90 degrees clockwise rotation matrix:\n", M_rotation_90_clockwise)
print("Matrix for the shear along x-axis:\n", M_shear_x)
```

```python
print("M_rotation_90_clockwise by M_shear_x:\n", M_rotation_90_clockwise @ M_shear_x)
print("M_shear_x by M_rotation_90_clockwise:\n", M_shear_x @ M_rotation_90_clockwise)
```
